## 使用實價登入資料

從政府所提供的開放資料中隨便找一個有趣的東西來玩

以下使用實價登入資料為例(雖然大家都知道實價登入都不是真實的)

為了要示範平面繪圖功能

所以我想要取兩個資料的維度來看

1. 先下載 JSON 資料

2. 使用程式讀入 JSON

3. 印出來看看有哪些資料


下面讀進來的 json 檔案放在變數 jf
因為 jf 很肥很大, 所以印出第零筆和第一筆

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import json

mpl.rcParams['font.sans-serif'] = ['FangSong']

with open('land/f_lvr_land_a.json' , 'r') as reader:
    jf = json.loads(reader.read())

print(jf[0])
print('\n')
print(jf[1])
```

## 整理資料

取 `鄉政市區` 和 `單價元平方公尺` 來玩玩看

如果要繪製長條圖

用來表示一個地區的均價

那麼我們要先把原始的資料分門別類

用一個大大的字典檔 `{ '天龍區': [123, 877, 978], '貧民窟':[20, 11, 3] }`

```python
dictPrice = {}

for row in jf[1:]:
    if row['單價元平方公尺']:
        if row['鄉鎮市區'] in dictPrice:
            dictPrice[row['鄉鎮市區']].append( int(row['單價元平方公尺']) )
        else:
            dictPrice[row['鄉鎮市區']] = [ int(row['單價元平方公尺']) ]

print(dictPrice)
```

```python
dicts = []  # 行政區
for d in dictPrice.keys():
    dicts.append(d)

dictsAvgPrice = []
for p in dictPrice.values():
    avg = sum(p)/len(p)
    dictsAvgPrice.append(avg)

for i in range(len(dicts)):
    print('{} --> {}'.format(dicts[i], dictsAvgPrice[i]))
    

plt.rcdefaults()
fig, ax = plt.subplots()

# Example data
y_pos = np.arange(len(dicts))
price = 100

ax.barh(y_pos, price, align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(dicts)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('平均交易單價(元/平方公尺)')
ax.set_title('炒房均價表')

plt.show()
```
